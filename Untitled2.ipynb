```python
import numpy as np
import pandas as pd
import time
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
# Função f(x) para soma de diferentes potenciações
def soma_diferentes_potencias(x):
    return sum(np.abs(x[i]) ** (i + 1) for i in range(len(x)))
```

```python
# Parâmetros
d = 2  # Dimensão da função (para visualização em 3D)
avaliacoes = 10000  # Número de avaliações sugeridas
limite_inf, limite_sup = -1, 1  # Limites de xi
```

```python
# Busca sequencial
def busca_sequencial(d, avaliacoes):
    melhor_valor = float('inf')
    melhor_x = None
    for i in range(avaliacoes):
        x = np.linspace(limite_inf, limite_sup, d) * (i / avaliacoes)
        valor = soma_diferentes_potencias(x)
        if valor < melhor_valor:
            melhor_valor = valor
            melhor_x = x
    return melhor_valor, melhor_x
```

```python
# Busca aleatória
def busca_aleatoria(d, avaliacoes):
    resultados = []
    for _ in range(10):  # 10 execuções independentes
        inicio = time.time()
        valores = []
        for _ in range(avaliacoes):
            x = np.random.uniform(limite_inf, limite_sup, d)
            valor = soma_diferentes_potencias(x)
            valores.append(valor)
        tempo_execucao = time.time() - inicio
        resultados.append({
            'melhor': min(valores),
            'pior': max(valores),
            'media': np.mean(valores),
            'mediana': np.median(valores),
            'desvio_padrao': np.std(valores),
            'tempo': tempo_execucao
        })
    return pd.DataFrame(resultados)
```

```python
# Executando busca sequencial
melhor_valor_sequencial, melhor_x_sequencial = busca_sequencial(d, avaliacoes)
```

```python
# Executando busca aleatória e gerando tabela de resultados
df_resultados = busca_aleatoria(d, avaliacoes)
```

```python
# Exibindo resultados em tabela
print(df_resultados)
```

```python
# Plotando gráfico da nuvem de pontos
x_vals, y_vals, z_vals = [], [], []
for _ in range(avaliacoes):
    x = np.random.uniform(limite_inf, limite_sup, d)
    x_vals.append(x[0])
    y_vals.append(x[1])
    z_vals.append(soma_diferentes_potencias(x))
```

```python
# Nuvem de pontos
plt.figure()
plt.scatter(x_vals, y_vals, c=z_vals, cmap='viridis', s=5)
plt.colorbar(label='f(x)')
plt.xlabel("x")
plt.ylabel("y")
plt.title("Nuvem de pontos para a função f(x)")
plt.scatter(melhor_x_sequencial[0], melhor_x_sequencial[1], color='red', label='Melhor ponto sequencial')
plt.legend()
plt.show()
```

```python
# Gráfico 3D
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(x_vals, y_vals, z_vals, c=z_vals, cmap='viridis', s=5)
ax.set_xlabel("x")
ax.set_ylabel("y")
ax.set_zlabel("f(x)")
ax.set_title("Gráfico 3D da função f(x)")
plt.show()
```

```python
# Gráfico 3D para Busca Sequencial
fig = plt.figure(figsize=(12, 6))

ax_seq = fig.add_subplot(121, projection='3d')
scatter_seq = ax_seq.scatter(x_vals, y_vals, z_vals, c=z_vals, cmap='viridis', s=0.6)
ax_seq.set_title("Busca Sequencial")
ax_seq.set_xlabel("x1")
ax_seq.set_ylabel("x2")
ax_seq.set_zlabel("Valor da Função")
fig.colorbar(scatter_seq, ax=ax_seq, label="Valor da Função")

# Gráfico 3D para Busca Aleatória
ax_rand = fig.add_subplot(122, projection='3d')
scatter_rand = ax_rand.scatter(x_vals, y_vals, z_vals, c=z_vals, cmap='plasma', s=0.6)
ax_rand.set_title("Busca Aleatória")
ax_rand.set_xlabel("x1")
ax_rand.set_ylabel("x2")
ax_rand.set_zlabel("Valor da Função")
fig.colorbar(scatter_rand, ax=ax_rand, label="Valor da Função")

plt.suptitle("Comparação de Avaliações da Função em Busca Sequencial e Aleatória (3D)")
plt.show()
```
